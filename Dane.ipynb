```
df = sns.load_dataset('tips')
df.head()
```

```
df.describe(include=['category'])
```

```
sns.relplot(data = df,x='total_bill',y='tip',hue='size',size='size',sizes=(10,200),palette='viridis',
            style='size')
```

### WYKRESY RELACYJNE

```
df = sns.load_dataset('fmri')
df.head(3)
```

```
sns.relplot(data=df, x='timepoint',y='signal',kind='line')
```

```
df.groupby('timepoint').agg('mean').plot()
```

```
sns.relplot(data=df, x='timepoint',y='signal',kind='line',hue='event',style='event',markers=True,col='region')
```

```
sns.relplot(data=df, x='timepoint',y='signal',kind='line',hue='event',style='event',markers=True,col='subject',col_wrap=4)
```

### CAT PLOT

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()
```

```
df = sns.load_dataset('tips')
df.head()
```

```
sns.set(style='ticks',palette='pastel')
sns.catplot(data=df,x='day',y='total_bill')
```

```
sns.catplot(data=df,x='day',y='total_bill',kind='boxen')
```

```
sns.catplot(data=df.query("day==['Fri','Sat','Sun']"),
            x='day',y='total_bill',kind='swarm', hue='sex')
```
