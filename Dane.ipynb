### CAT PLOT

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()
```

```
df = sns.load_dataset('tips')
df.head()
```

```
sns.set(style='ticks',palette='pastel')
sns.catplot(data=df,x='day',y='total_bill')
```

```
sns.catplot(data=df,x='day',y='total_bill',kind='boxen')
```

```
sns.catplot(data=df.query("day==['Fri','Sat','Sun']"),
            x='day',y='total_bill',kind='swarm', hue='sex')
```

### WYKRESY ROZKADW

```
df = np.random.multivariate_normal(mean=[0,1], cov=[(1,-0.4),(-0.4,1)],size=200)
df= pd.DataFrame(df,columns=['x','y'])
df.head()
```

```
df.describe()
```

```
sns.set(style='white')
sns.jointplot(data=df, x='x', y='y',kind='hex')
```

```

g=sns.jointplot(data=df, x='x', y='y',kind='kde',n_levels=15,cmap='viridis')

```

### WYKRESY RELACJI PARAMI PAIRPLOT

```
df = sns.load_dataset('iris')
df.head(3)
```

```
sns.pairplot(data=df, hue='species')
```

```
g = sns.PairGrid(df, hue ='species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6)
```
